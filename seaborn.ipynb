## API
* Seaborn имеет 2 API (figure and axes level functions)
* `axes` - возвращает объект `Axes` из `matplotlib`
* `figure` - возвращает `FaceGrid`, в котором можно нарисовать несколько графиков

```python
ax = sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins)
```

```python
type(ax)
```

```python
fg = sns.relplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, kind='scatter')
```

## Join plot и Pair plot
* Join plot - совместное распределение 2 переменных и распределение каждой переменной отдельно
* Pair plot - попарное распределение всех переменных и распределение каждой переменной отдельно

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
sns.pairplot(data=penguins, hue="species")
```
